```python
# cideries by cluster

# Examine cluster 0

def cidery_stats(df):
    # get number of cideries per city 
    cideries_no = df.Frequency.sum()
    
    # get cideries per city
    cideries_per_city = cideries_no / len(df.Frequency)
    
    # get percent of cities with cidery 
    cities_with_cideries = 0 
    for i in df.Frequency:
        if i > 0:
            cities_with_cideries += 1 
            
    cideries_pct = cities_with_cideries / len(df.Frequency)
    
    return cideries_no, cideries_per_city, cideries_pct

print(cidery_stats(city_cluster_0))
print(cidery_stats(city_cluster_1))
print(cidery_stats(city_cluster_2))


# cideries_0 = city_cluster_0.Frequency.sum()
# cideries_per_city_0 = cideries_0 / len(city_cluster_0.Frequency)

# print(cideries_per_city_0)



# cideries_1 = city_cluster_1.Frequency.sum()
# cideries_per_city_1 = cideries_0 / len(city_cluster_0.Frequency)


# cideries_2 = city_cluster_2.Frequency.sum()
city_cluster_2
```
