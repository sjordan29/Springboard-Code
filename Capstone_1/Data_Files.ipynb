```python
# I need to rename the columns 

income_column_names = pd.read_csv('data/ACS_16_1YR_S1901_metadata.csv', header=None)
income_column_names.head()

# Need to get rid of punctuation and spaces

punctuation = '''!()-[]{};:'"\,<>./?@#$%^&*~'''

income_column_names_list = []
import string

for entry in income_column_names[1]:
    new_entry = ''
    for char in entry:
        if char == ' ':
            char = '_'
        if char not in punctuation:
                new_entry = new_entry + char
        
    income_column_names_list.append(new_entry)
    
# Replace column names

income.columns = income_column_names_list

income.head()   
```

```python
age = pd.read_csv('data/ACS_16_1YR_S0101_with_ann.csv')
age.head()
```

```python
# I need to rename the columns 

age_column_names = pd.read_csv('data/ACS_16_1YR_S0101_metadata.csv', header=None)
age_column_names.head()

# Need to get rid of punctuation and spaces

punctuation = '''!()-[]{};:'"\,<>./?@#$%^&*~'''

age_column_names_list = []

import string

for entry in age_column_names[1]:
    new_entry = ''
    for char in entry:
        if char == ' ':
            char = '_'
        if char not in punctuation:
                new_entry = new_entry + char
        
    age_column_names_list.append(new_entry)
    
# Replace column names

age.columns = age_column_names_list

age.head()
```

```python
population = pd.read_csv('data/ACS_16_1YR_B01003_with_ann.csv')
population.head()
```

```python
# Need to rename columns
population.columns = ['Id', 'Id2', 'Geo_Display_Label', 'Population', 'Population_Margin_of_Error']
population.head()
```

```python
age_and_income = pd.merge(age, income)
demographics = pd.merge(age_and_income, population)
demographics.head()
```

```python
# I need to have the state of each city in a separate column so that I can join cider data with demographic data 
demographics[['City', 'State']] = demographics['Geography'].str.split(',', expand=True)
demographics.head()
```

```python
cider_data = pd.ExcelFile('data/cider_data.xlsx')
state_shipments = cider_data.parse('Sheet1')
state_volume = cider_data.parse('Page1')

state_shipments.columns = ['State', 'Shipments_16']

# Notice that there is a leading space in the State column of demographics 

print(demographics['State'][0])
print(state_shipments['State'][0])

# Get rid of the leading space
    
for i, row in demographics.iterrows():
    demographics.loc[i, 'State'] = demographics['State'][i].lstrip(' ')

# Now the dataframes can be merged on the State column 

dems_and_cider = demographics.merge(state_shipments, on = 'State', how= 'outer')

# Let's make sure that the data has been filled in appropriately
print(dems_and_cider.Shipments_16.unique())

# There are some values of nan. Let's investigate.

print(dems_and_cider.loc[dems_and_cider['Shipments_16'].isnull()])

# There's only one NaN at the end

dems_and_cider.tail()
```
