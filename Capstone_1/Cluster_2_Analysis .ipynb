```python
import pandas as pd

# Add region data to help break down the analysis 

target_cities = pd.read_csv('data/cidery_cluster_2.csv')
state_regions = pd.read_csv('data/State_Regions.csv', header = None)
state_regions.columns = ['State', 'Region']
state_regions

target_cities = pd.merge(target_cities, state_regions, how='left')
target_cities = target_cities.sort_values('Region')
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

dims = (14, 5)
fig, ax = plt.subplots(figsize=dims)


sns.set_style("whitegrid")
ax = sns.barplot(x = "City", y = "Total_Estimate_SUMMARY_INDICATORS__Median_age_years", 
                    data = target_cities, hue = "Region")

for item in ax.get_xticklabels():
    item.set_rotation(80)
    
ax.set(xlabel='City', ylabel = 'Median Age (yr)')

def change_width(ax, new_value):
    for patch in ax.patches:
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
    
        
change_width(ax, 0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)


plt.show()
```
