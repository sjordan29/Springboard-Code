```python
# import modules 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# import data
paris = pd.read_csv('paris.csv')
```

```python
paris.head()
```

```python
sns.barplot(x='accommodates', y='price', data=paris)
plt.ylabel('Mean Price')
plt.xlabel('Rental Capacity (# of people)')
plt.title('Mean Price by Rental Capacity')
plt.show()

sns.violinplot(paris.accommodates, paris.price)
plt.ylabel('Price')
plt.xlabel('Rental Capacity (# of people)')
plt.title('Price by Rental Capacity')
plt.show()

log_price = np.log10(paris.price)

sns.violinplot(paris.accommodates, log_price)
plt.ylabel('Log(Price)')
plt.xlabel('Rental Capacity (# of people)')
plt.title('Log(Price) by Rental Capacity')
plt.show()
```

```python
# plot ecdf of price
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)
    # x-data for the ECDF: x
    x = np.sort(data)
    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n
    return x,y

x, y = ecdf(paris.price)

plt.plot(x, y, marker='.', linestyle='none')
plt.xlabel('Rental Price per Night')
plt.ylabel('ECDF')
plt.show()
```

```python
x_log, y_log = ecdf(log_price)

plt.plot(x_log, y_log, marker='.', linestyle='none')
plt.xlabel('Log of Rental Price per Night')
plt.ylabel('ECDF')
plt.show()
```

```python
plt.hist(log_price, bins = 20)
plt.xlabel('Log of Price per night')
plt.ylabel('Count')
plt.show()
```

The price data is very skewed, but looking at log of price has a more normal distribution.

```python
paris.loc[paris['neighborhood'] == 'Plaine-Monceau', 'arrondissement'] = 17
paris.loc[paris['neighborhood'] == 'Saint-Avoye', 'arrondissement'] = 3
paris.loc[paris['neighborhood'] == 'Goutte-dOr', 'arrondissement'] = 18
paris.loc[paris['neighborhood'] == 'Saint-Thomas-dAquin', 'arrondissement'] = 7
paris.loc[paris['neighborhood'] == 'Chaussée-dAntin', 'arrondissement'] = 9
paris.loc[paris['neighborhood'] == 'St-Germain-lAuxerrois', 'arrondissement'] = 1
```

```python
sns.barplot(x='arrondissement', y='price', data=paris)
plt.xlabel('Arrondissement')
# plt.ylabel('Mean Price')
plt.title('Mean Price by Arrondissement')
plt.show()

sns.violinplot(paris.arrondissement, paris.price)
plt.xlabel('Arrondissement')
# plt.ylabel('Price')
plt.title('Price by Arrondissement')
plt.show()

sns.violinplot(paris.arrondissement, log_price)
plt.xlabel('Arrondissement')
# plt.ylabel('Log Price')
plt.title('Log of Price by Arrondissement')
plt.show()
```

Rentals in arrondissements 1-8 have the highest average prices - this makes sense, as they are in the center of the city near many major attractions (the Louvre, Eiffel Tower, Champs-Élysées, Arc de Triomphe, etc).
