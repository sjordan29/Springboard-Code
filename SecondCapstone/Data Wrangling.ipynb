```python
def add_city(df, city):
    df['city'] = city
    return df 

add_city(chicago_3, 'Chicago')
add_city(philly_3, 'Philadelphia')
add_city(minneapolis_3, 'Minneapolis') 

all_cities = pd.concat([chicago_3, philly_3, minneapolis_3])
```

```python
all_cities = all_cities.reset_index()

all_cities = all_cities.drop(['no_buildings'], axis=1)
```

```python
# all_cities['energy_use_2014'] = all_cities['eui_2014'] * all_cities['gba']
# all_cities['energy_use_2015'] = all_cities['eui_2015'] * all_cities['gba']
# all_cities['energy_use_2016'] = all_cities['eui_2016'] * all_cities['gba']

cols = ['energy_star_score_2014', 'energy_star_score_2015', 'energy_star_score_2016', 'eui_2014', 'eui_2015', 
       'eui_2016', 'gba', 'year_built']

for col in cols:
    all_cities[col] = pd.to_numeric(all_cities[col], errors = 'coerce')
    
all_cities.round(1)
    

all_cities['year_built'] = pd.to_numeric(all_cities['year_built'], downcast = 'integer')
all_cities=all_cities.drop('index', axis=1)
all_cities.info()
all_cities.head()
```

```python
all_cities.describe()
```

```python
# there are some null values in the eui_2014 column that slipped through the cracks
# same for a few in 2015 and 2016

all_cities.loc[(all_cities['eui_2014'].isnull() == True)]

# clearly there were some non-null values in eui for 2014
# examination shows that these were not marked as null originally but rather as 'Not Available'
# Let's get rid of those

all_cities = all_cities.loc[(all_cities['eui_2016'].isnull()== False) &(all_cities['eui_2015'].isnull()== False) &
                           (all_cities['eui_2014'].isnull()== False)]

all_cities.describe()
```

```python
all_cities.reindex()

# there is an outlier for eui_2016 - the max value is WAY too high 
all_cities.loc[all_cities['eui_2016'] == 29596.6]

# this is clearly a mistake in data entry. I am going to drop the row.
all_cities = all_cities.loc[all_cities['eui_2016'] != 29596.6]
```

```python
all_cities.describe()

# some values still seem very high 
all_cities.loc[all_cities['eui_2016'] > 800]

# the parking garage with an eui of 9269 kbtus/sq ft (and a huge variation in eui year over year) clearly has data issues
# i will remove 

all_cities = all_cities.loc[all_cities['eui_2016'] < 9000]

all_cities.describe()
```
